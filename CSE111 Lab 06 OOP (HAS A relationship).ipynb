```python
,#Task 1 (Teacher with the Courses)
class Teacher:
  def __init__(self,teacher_name,dept):
    self.teacher_name = teacher_name
    self.dept         = dept
    self.courses      = []

  def addCourse(self,add_course):
    self.courses.append(add_course.course_name)
  def printDetail(self):
    print("====================================")
    print(f"Name:  {self.teacher_name}\nDepartment:  {self.dept}\nList of courses")
    print("====================================")
    for i in self.courses:
      print(i)
    print("====================================")
class Course:
  def __init__(self,course_name):
    self.course_name = course_name
t1 = Teacher("Saad Abdullah", "CSE")
t2 = Teacher("Mumit Khan", "CSE")
t3 = Teacher("Sadia Kazi", "CSE")
c1 = Course("CSE 110 Programming Language I")
c2 = Course("CSE 111 Programming Language-II")
c3 = Course("CSE 220 Data Structures")
c4 = Course("CSE 221 Algorithms")
c5 = Course("CSE 230 Discrete Mathematics")
c6 = Course("CSE 310 Object Oriented Programming")
c7 = Course("CSE 320 Data Communications")
c8 = Course("CSE 340 Computer Architecture")
t1.addCourse(c1)
t1.addCourse(c2)
t2.addCourse(c3)
t2.addCourse(c4)
t2.addCourse(c5)
t3.addCourse(c6)
t3.addCourse(c7)
t3.addCourse(c8)
t1.printDetail()
t2.printDetail()
t3.printDetail()
```

```python
#TASK 2 (Student Details)
class Student:
    def __init__(self, name, student_id, cgpa):
        self.name = name
        self.student_id = student_id
        self.cgpa = cgpa
    def set_id(self, student_id):
        self.student_id = student_id
class Department:
    def __init__(self, dept):
        self.dept = dept
        self.students = []
    def add_student(self, *students):
      for student in students:
        exists = False
        for s in self.students:
          if student.student_id == s.student_id:
            exists = True
            break
        if exists == True:
          print("Student with the same ID already exists, Please try with another ID")
        else:
          self.students.append(student)
          print(f"Welcome to {self.dept} department, {student.name}")
    def find_student(self, id):
        found = False
        for student in self.students:
            if student.student_id == id :
                print(f"Student info:\nStudent Name: {student.name}\nID: {student.student_id}\nCGPA: {student.cgpa}")
                found = True
        if  found == False:
            print("Student with this ID doesn't exist, Please give a valid ID")
    def details(self):
        print(f"Department Name: {self.dept}")
        print(f"Number of student: {len(self.students)}")
        print("Details of the students: ")
        for student in self.students:
            print(f"Student name: {student.name}, ID: {student.student_id}, cgpa: {student.cgpa}")

s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.find_student(22112233)
print("2==================================")
cse.add_student(s1, s2, s3)
print("3==================================")
cse.details()
print("4==================================")
cse.find_student(22301010)
print("5==================================")
s4 = Student("Nakib", 22301010, 3.22)
cse.add_student(s4)
print("6==================================")
s4.set_id(21201220)
cse.add_student(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib", 22201010, 2.29)
cse.add_student(s5)
print("9==================================")
cse.details()
```

```python
# TASK 3 (Vaccine)
class Vaccine:
  def __init__(self,name_vaccine,country,days):
    self.name_vaccine = name_vaccine
    self.country = country
    self.days = days

class Person:
  def __init__(self,name ,age, profession = None):
    self.name = name
    self.age = age
    self.profession = profession
    self.count = []
    self.days = []
    self.vaccine_name = []


  def pushVaccine(self,vaccine,dose_count = "1st Dose"):
    if dose_count == "1st Dose" :
      if self.age >= 25 :
        print(f"1st dose done for {self.name}")
        self.days.append(vaccine.days)
        self.count.append(dose_count)
        self.vaccine_name.append(vaccine.name_vaccine)
      elif self.profession == "Student":
        print(f"1st dose done for {self.name}")
        self.days.append(vaccine.days)
        self.count.append(dose_count)
        self.vaccine_name.append(vaccine.name_vaccine)
      else:
        print(f"Sorry {self.name}, Minimum age for taking vaccines is 25 years now")
    elif dose_count == "2nd Dose":
      if vaccine.name_vaccine in self.vaccine_name:
        print(f"2nd dose done for {self.name}")
        self.count.append(dose_count)
      else:
        print(f"Sorry {self.name}, you can’t take 2 different vaccines")

  def showDetail(self):
    print(f"Name: {self.name} Age: {self.age} Type: {self.profession}")
    print(f"Vaccine name: {self.vaccine_name[0]}")
    if len(self.count) == 1 :
      print(f"1st dose: Given")
      print(f"2nd dose: Please come after {self.days[0]} days")
    elif len(self.count) == 2 :
      print("1st dose: Given")
      print("2nd dose: Given")


astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
