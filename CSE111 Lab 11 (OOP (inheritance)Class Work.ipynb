```python
#Class work 2
import math
class Point:
  def __init__(self, x=0, y=0):
    self.x = x
    self.y = y
    self.area = 0
  def calculate_area(self):
    return round(self.area,4)
  def print_details(self):
    print("--------- Printing details ----------")
    print(f'Co-ordinate: ({self.x},{self.y})')
    print(f'Area: {self.area}')

class Circle(Point):
  def __init__(self,radius ,x = None, y = None):
    super().__init__(x,y)
    self.radius = radius
  def  calculate_area(self):
    return round(math.pi * (self.radius **2),4)
  def  print_details(self):
    print("--------- Printing details ----------")
    print(f'Co-ordinate: ({self.x},{self.y})')
    print(f'Area: {round(math.pi * (self.radius **2),4)}')
    print(f"Radius: {self.radius}")
class Sphere(Circle):
  def __init__(self,radius ,x = None, y = None):
    super().__init__(x,y)
    self.radius = radius
  def  calculate_area(self):
    return round(4*math.pi * (self.radius **2),4)
  def  print_details(self):
    print("--------- Printing details ----------")
    print(f'Co-ordinate: ({self.x},{self.y})')
    print(f'Area: {round(4*math.pi * (self.radius **2),4)}')
    print(f"Radius: {self.radius}")

print("--------------1---------------")
p1 = Point(2,3)
print(f'Area of p1: {p1.calculate_area()}')
print("--------------2---------------")
p1.print_details()
print("--------------3---------------")
p2 = Point()
p2.print_details()
print("--------------4---------------")
c1 = Circle(4,0,3)
print(f'Area of c1: {c1.calculate_area()}')
print("--------------5---------------")
c1.print_details()
print("--------------6---------------")
c2 = Circle( 7)
print(f'Area of c2: {c2.calculate_area()}')
print("--------------7---------------")
sph1 = Sphere(3,0,2)
print(f'Area of sph1: {sph1.calculate_area()}')
print("--------------8---------------")
sph1.print_details()
print("--------------9---------------")
sph2 = Sphere(6)
print(f'Area of sph2: {sph2.calculate_area()}')
```

```python
#Class work 3
class Account:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance
        self.account_type = "General"
        self.maturity = 0

    def print_details(self):
        print("------ Account details ------")
        print(f"Account Type: {self.account_type}, Maturity: {self.maturity} years")
        print(f"Account Number: {self.account_number}, Balance: ${self.balance:.2f}")

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount:.2f}. New Balance: ${self.balance:.2f}")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrew ${amount:.2f}. New Balance: ${self.balance:.2f}")
        else:
            print("Insufficient funds.")

    def year_passed(self, year):
        self.maturity += year
        print(f"Maturity of the account: {self.maturity} years")


class SavingsAccount(Account):
    def __init__(self, account_type, account_number, balance, interest_rate, minimum_limit):
        super().__init__(account_number, balance)
        self.account_type = "Savings"
        self.balance =  balance
        self.interest_rate = interest_rate
        self.minimum_limit = minimum_limit
    def withdraw(self, amount):
      if self.minimum_limit <= self.balance - amount :
        super().withdraw(amount)
      else:
        print("Insufficient funds.")


    def print_details(self):
        print(f"Account Type: {self.account_type}, Maturity: 0 years\n"
              f"Account Number: {self.account_number}, Balance: ${self.balance}\n"
              f"Interest Rate: {self.interest_rate}, Minimum Limit: {self.minimum_limit}")

    def apply_interest(self):
        interest_amount = self.balance * self.interest_rate
        self.balance += interest_amount
        print(f"Interest applied. New Balance: ${self.balance:.2f}")


class FixedDepositAccount(Account):
    def __init__(self, account_type, account_number, balance, maturity_years):
        super().__init__(account_number, balance)
        self.account_type = "Fixed Deposit"
        self.maturity_years = maturity_years

    def print_details(self):
        print(f"Account Type: {self.account_type}, Maturity: {self.maturity_years} years\n"
              f"Account Number: {self.account_number}, Balance: ${self.balance:.2f}")

    def deposit(self, amount):
        print("You cannot deposit into a fixed deposit account.")

    def withdraw(self, amount):
        if self.maturity >= self.maturity_years:
            super().withdraw(amount)
            self.maturity_years = self.maturity

        else:
            print("Cannot withdraw, Account is not matured")


print("-----------1------------")
account = Account("A203", 2000)
account.print_details()
print("-----------2------------")
account.deposit(400)
account.withdraw(1500)
account.year_passed(2)
print("-----------3------------")
account.print_details()
print("-----------4------------")
savings_account = SavingsAccount("Savings", "SA123", 1000, 0.05, 500)
savings_account.print_details()
print("-----------5------------")
savings_account.deposit(400)
print("-----------6------------")
savings_account.withdraw(1000)
print("-----------7------------")
savings_account.withdraw(800)
print("-----------8------------")
savings_account.apply_interest()
print("-----------9------------")
savings_account.print_details()
print("-----------10------------")
fixed_account1 = FixedDepositAccount("Fixed Deposit", "FDA321", 10000, 5)
fixed_account1.print_details()
print("-----------11------------")
fixed_account1.deposit(400)
print("-----------12------------")
fixed_account1.year_passed(6)
print("-----------13------------")
fixed_account1.withdraw(10000)
print("-----------14------------")
fixed_account1.print_details()
print("-----------15------------")
fixed_account2 = FixedDepositAccount("Fixed Deposit", "FDA300", 50000, 7)
fixed_account2.print_details()
print("-----------16------------")
fixed_account2.withdraw(10000)
```
