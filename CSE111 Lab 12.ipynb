```python
# Practice Sheet TASK - 9
class Fruit:
    Total_order = 0

    def __init__(self, Order_ID, weight):
        self.Order_ID = Order_ID
        self.weight = weight
        Fruit.Total_order = Fruit.Total_order + 1

    def __str__(self):
        return f"Order Id {self.Order_ID}, Weight: {self.weight},"


class Mango(Fruit):
    def __init__(self, id, weight, name, price):
        super().__init__(id, weight)
        self.name = name
        self.price = price

    def __str__(self):
        total_price = self.weight * self.price
        return f"{super().__str__()} Variety: {self.name},\nTotal Price: {total_price}"

    def __add__(self, other):
        total_weight = self.weight + other.weight
        total_price = (self.weight * self.price) + (other.weight * other.price)
        return Mango("Combined Order", total_weight, "Combined Mango", total_price)


class JackFruit(Fruit):
    def __init__(self, id, weight, price):
        super().__init__(id, weight)
        self.price = price

    def __str__(self):
        total_price = self.weight * self.price
        return f"{super().__str__()} Total Price: {total_price}"

    def __add__(self, other):
        total_weight = self.weight + other.weight
        total_price = (self.weight * self.price) + (other.weight * other.price)
        return JackFruit("Combined Order", total_weight, total_price)


m1 = Mango(1, 5, "GopalVog", 250)
print(m1)
m2 = Mango(2, 5, "HariVanga", 230)
print(m2)
j1 = JackFruit(3, 5, 250)
print(j1)
j2 = JackFruit(4, 4, 210)
print(j2)
print("Total number of Orders:", Fruit.Total_order)
print("==================")
print("The total Price of the orders are:", (m1 + m2).price)
print("==================")
print("The total Price of the orders are:", (j1 + j2).price)
```

```python
# Practice Sheet TASK - 18
class User:
    def __init__(self, name, address, ride_type="single"):
        self.name = name
        self.address = address
        self.ride_type = ride_type
        self.status = False
        self.taxi_number = ""

    def check_status(self):
        if not self.status:
            print(f"Status: {self.name} is looking for a {self.ride_type}!")
        else:
            print(f"Status: {self.name} boarded in car {self.taxi_number}!")

class Uber:
    def __init__(self, taxi_number, ride_type, *route):
        self.taxi_number = taxi_number
        self.ride_type = ride_type
        self.route = list(route)

    def details(self):
        print(f"Car number: {self.taxi_number}")
        print(f"Type: {self.ride_type}")
        print(f"Routes: {' --> '.join(self.route)}")

    def pick(self, *users):
        for user in users:
            if user.address in self.route:
                if user.ride_type == self.ride_type:
                    user.status = True
                    user.taxi_number = self.taxi_number
                    print(f"Status: {user.name} boarded in car {self.taxi_number} !")
                else:
                    print(f"Status: {user.name} is looking for a {user.ride_type} ride !")
            else:
                print(f"Status: {user.name} is looking for a different route!")

# Creating users
user1 = User("Brooks", "Banani", "Shared")
user2 = User("Jocelyn", "Uttara")
user3 = User("Robert", "Gulshan", "Shared")
user4 = User("Langdon", "Mohakhali", "Shared")

# Displaying initial status of users
user1.check_status()
user2.check_status()
user3.check_status()
user4.check_status()
print("----------------------------------")

# Creating an Uber car instance
car1 = Uber("0K32BH", "Shared", "Mohakhali", "Banani", "Nikunja", "Uttara")
car1.details()
print("----------------------------------")

# Picking up users
car1.pick(user1, user2, user3, user4)
print("----------------------------------")

# Displaying status after picking up users
user1.check_status()
user2.check_status()
user3.check_status()
user4.check_status()
print("----------------------------------")

# Creating more Uber car instances
car2 = Uber("5GD2BD", "Single", "Uttara")
car3 = Uber("4T12FR", "Shared", "Gulshan", "Bashundhara")
car2.details()
car3.details()
print("----------------------------------")

# Picking up users with the new cars
car2.pick(user2, user3)
print("----------------------------------")
car3.pick(user3)
print("----------------------------------")

# Displaying status after picking up more users
user2.check_status()
user3.check_status()
```
