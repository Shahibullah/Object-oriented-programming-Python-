```python
#TASK 1
class Student:
  def __init__(self,name,cgpa,credit = 9,dept = "CSE"):
    self.name = name
    self.cgpa = cgpa
    self.credit = credit
    self.dept = dept
    self.x = ""

  def checkScholarshipEligibility(self):
    if (self.cgpa >= 3.5) and (self.credit>10):
      if (3.7 >= self.cgpa >= 3.5):
        self.x = (f"{self.name} is eligible for Need-based scholarship.")
        print(self.x)
        return self.x

      else:
        self.x = (f"{self.name} is eligible for Merit-based scholarship.")
        print(self.x)
        return self.x
    else:
      self.x = (f"{self.name} is not eligible for scholarship.")
      print(self.x)
      return self.x
  def showDetails(self):
    print("Name: ",self.name)
    print("Department: ",self.dept)
    print("CGPA: ",self.cgpa)
    print("Number of Credits:",self.credit)
    print("Scholarship Status: ",self.x)

print('--------------------------')
std1 = Student("Alif", 3.99, 12)
print('--------------------------')
std1.checkScholarshipEligibility()
print('--------------------------')
std1.showDetails()
print('--------------------------')
std2 = Student("Mim", 3.4)
std3 = Student("Henry", 3.5, 15,"BBA")
print('--------------------------')
std2.checkScholarshipEligibility()
print('--------------------------')
std3.checkScholarshipEligibility()
print('--------------------------')
std2.showDetails()
print('--------------------------')
std3.showDetails()
print('--------------------------')
std4 = Student("Bob", 4.0, 6, "CSE")
print('--------------------------')
std4.checkScholarshipEligibility()
print('--------------------------')
std4.showDetails()
```

```python
#TASK 2
menu = {'Chicken Lollipop':15,'Beef Nugget':20,'Americano':180,'Red Velvet':150,'Prawn Tempura':80,'Saute Veg':200}
class Foodie:
  def __init__(self,name):
    self.name = name
    self.customer_order = {}
    self.cost = 0
    self.product = []
    self.tips = 0
  def show_orders(self):
    return (f"{self.name} has {len(self.product)} item(s) in the cart.\nItems: {self.product}\nTotal spent {self.cost+self.tips} .")
  def order(self,*args):
    list1 = []
    items = []
    quantity = []
    for i in range (len(args)):
      list1 = args[i].split("-")
      items.append(list1[0])
      quantity.append(int(list1[1]))
    self.product += items
    for i in range (len(items)):
      print(f"Ordered - {items[i]}, quantity - {quantity[i]},")
      for key , val in menu.items():
        if key == items[i]:
          print(f"price (per Unit) - {val}.")
          print(f"Total price - {val*quantity[i]}")
          self.cost += val*quantity[i]
  def pay_tips(self,tips = 0):
    self.tips = tips
    if self.tips == 0 :
      print("No tips to the waiter.")
    else:
      print(f"Gives {self.tips}/- tips to the waiter.")
f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3','Beef Nugget-6','Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())
f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6','Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())
```

```python
#TASK 3
class Department:
  def __init__(self,dept = "ChE",sections = 5):
    self.dept = dept
    self.sections = sections
    self.total_students = 0
    self.average_students = 0


    print(f"The {self.dept} Department has {self.sections} sections.")

  def add_students(self,*args):
    if (len(args)) < self.sections:
      print(f"The {self.dept} Department doesn't have 3 sections.")
    else:
      for i in args:
        self.total_students += i
      self.average_students = self.total_students / self.sections

      print(f"The {self.dept} Department has an average of {self.average_students:.2f} students in each section.")

  def merge_Department(self,*other_dept):
    for department in other_dept:
      print(f"{department.dept} Department is merged to {self.dept} Department.")
      self.total_students += department.average_students * department.sections
    self.average_students = self.total_students / self.sections
    return (f"Now the {self.dept} has an average of {self.average_students:.2f} students in each section.")
d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22,
32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,
51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
#TASK - 4
class Shopidify:
  def __init__(self,customer_name = None):
    if customer_name == None:
      self.customer_name = "Guest"
      print("Welcome to Shopidify")
    else:
      self.customer_name = customer_name
      print(f"Welcome {self.customer_name} to Shopidify")
    self.product_list = []
    self.cart = {}
    self.quantity = []
    self.transaction = 0

  def add_to_cart(self,*args):
    if len(args)  == 2 :
      product_name = args[0]
      quantity = args[1]
      self.product_list.append(product_name)
      self.quantity.append(quantity)
    elif len(args)  == 1 and type(args[0]) == str :
      product_name = args[0]
      quantity = 1
      self.product_list.append(product_name)
      self.quantity.append(quantity)
    elif len(args)  == 1 and type(args[0]) == list :
      for i in range (0,len(args[0]),2):
        self.product_list.append(args[0][i])
        self.quantity.append(args[0][i+1])
  def display_cart(self):
    print(f"Items in the cart for {self.customer_name}:")
    for i in  range (len(self.product_list)):
       if self.product_list[i] not in self.cart:
         self.cart.update({self.product_list[i]:self.quantity[i]})
       else:
         self.cart.update({self.product_list[i]:self.cart[self.product_list[i]]+self.quantity[i]})


    self.product_list = []
    self.quantity = []
    for k,v in  self.cart.items():
      print(f"- {k}: {v}x")
  def checkout(self):
    print(f"Checkout completed for {self.customer_name}")
  def display_history(self):
    self.transaction += 1
    print(f"Purchase history for {self.customer_name}:")
    print(f"Transaction {self.transaction}:")
    for k,v in  self.cart.items():
      print(f"- {k}: {v}x")

guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
#TASK 5
class Author:
    def __init__(self, name=None):
        self.name = name
        self.books = {}

    def setName(self, name):
        self.name = name

    def addBook(self, title, genre):
        if not self.name:
            print("A book can not be added without author name")
        else:
            if genre not in self.books:
                self.books[genre] = [title]
            else:
                self.books[genre].append(title)

    def printDetail(self):
        if not self.name:
            print("Error: Author name is missing.")
        else:
            print("Number of Book(s):", sum(len(books) for books in self.books.values()))
            print("Author Name:", self.name)
            for genre, titles in self.books.items():
                print(f"{genre}: {', '.join(titles)}")


# Test cases
print("=================================")
a1 = Author()
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```
