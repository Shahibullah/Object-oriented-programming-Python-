```python
#TASK 1
class NikeBangladesh:
    branches_opened = []

    def __init__(self, location):
        self.location = location
        self.shop_items = {"Air Jordan": 0, "Cortez": 0, "Zoom Kobe": 0}
        self.sold = 0
        NikeBangladesh.branches_opened.append(self)

    def restockProducts(self, products):
        for product, quantity in products.items():
            self.shop_items[product] += quantity

    def productSold(self, products):
        for product, quantity in products.items():
            self.shop_items[product] -= quantity
            self.sold += quantity

    def details(self):
        print(f"Nike {self.location} outlet:")
        print("Products Currently Stocked: ")
        print(self.shop_items)
        print(f"Sold: {self.sold}")

    @classmethod
    def status(cls):
        print("Nike Bangladesh Status:")
        temp = []

        for branch in cls.branches_opened:
          temp.append(branch.location)
        print(f"Branches Opened:{temp}")

        print("Currently Stocked")

        total_stock = {"Air Jordan": 0, "Cortez": 0, "Zoom Kobe": 0}
        total_sold = 0
        for branch in cls.branches_opened:
            for product, quantity in branch.shop_items.items():
                total_stock[product] += quantity
            total_sold += branch.sold
        print(total_stock)
        print(f"Sold: {total_sold}")


print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts(
    {"Air Jordan": 1200, "Cortez": 200, "Zoom Kobe": 200})
chittagong.restockProducts(
    {"Air Jordan": 1000, "Cortez": 250, "Zoom Kobe": 100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan": 760, "Cortez": 90})
chittagong.productSold({"Air Jordan": 520, "Zoom Kobe": 70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```

```python
#TASK 2
class Player:
    total = 0
    player_list = []

    def __init__(self, name = "Nai", number=10, team=None):
        self.name = name
        self.number = number
        self.team = team
        if name != "Nai" :
            Player.total += 1
            Player.player_list.append(name)

    def set_name(self, name):
        self.name = name
        Player.total += 1
        Player.player_list.append(name)

    def set_number(self, number):
        self.number = number

    def set_team(self, team):
        self.team = team

    def player_detail(self):
        return f"Player Name: {self.name}\nJersey Number: {self.number}\nCountry: {self.team}"

    @classmethod
    def info(cls):
        print("Total number of players:", cls.total)
        print("Players enlisted so far:", ', '.join(cls.player_list))



print("Total number of players:", Player.total)
print("---------------------------")
p1 = Player()
p1.set_name("Neymar")
p1.set_team("Brazil")
print(p1.player_detail())
print('========================')
Player.info()
print("---------------------------")
p2 = Player("Ronaldo")
p2.set_number(7)
p2.set_team("Portugal")
print(p2.player_detail())
print('========================')
Player.info()
print("---------------------------")
p3 = Player("Messi")
p3.set_team("Argentina")
print(p3.player_detail())
print('========================')
Player.info()
print("---------------------------")
p4 = Player("Mbappe", 10, "France")
print(p4.player_detail())
print('========================')
Player.info()
```
